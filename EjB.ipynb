```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# plt.figure(figsize=(10, 6))
```

```python
# 1. Equation 1: dy/dx + 2y/x = sin(x)/x^2
def eq1(y, x):
    return (np.sin(x) / x**2) - (2 * y / x)

y0_1 = 0
x_vals_1 = np.linspace(0.1, 10, 100)
y_vals_1 = odeint(eq1, y0_1, x_vals_1)

# Plot
plt.figure()
plt.subplot(3, 1, 1)
plt.plot(x_vals_1, y_vals_1, label="Equation 1: dy/dx + 2y/x = sin(x)/x^2", color='b')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
```

```python
# 2. Equation 2: dy/dx - y/x = -x * e^(-x)
def eq2(y, x):
    return -x * np.exp(-x) + y / x

y0_2 = 1
x_vals_2 = np.linspace(0.1, 10, 100)
y_vals_2 = odeint(eq2, y0_2, x_vals_2)

# Plot
plt.figure()
plt.subplot(3, 1, 2)
plt.plot(x_vals_2, y_vals_2, label="Equation 2: dy/dx - y/x = -x * e^(-x)", color='g')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
```

```python
# 3. Equation 3: dy/dx + 2xy = x
def eq3(y, x):
    return x - 2 * x * y

y0_3 = 0
x_vals_3 = np.linspace(0, 10, 100)
y_vals_3 = odeint(eq3, y0_3, x_vals_3)

# Plot
plt.figure()
plt.subplot(3, 1, 3)
plt.plot(x_vals_3, y_vals_3, label="Equation 3: dy/dx + 2xy = x", color='r')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
```
