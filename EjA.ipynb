```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# 1. Equation 1: dy/dx - e^(-y) * sec^2(x) = 0, rewritten as dy/dx = e^(-y) * sec^2(x)
def eq1(y, x):
    return np.exp(-y) * (1/np.cos(x))**2

y0_1 = 0 
# Values of x (avoiding x = pi/2, where sec(x) -> infinity)
x_vals = np.linspace(0, np.pi/2 - 0.01, 100)
y_vals_1 = odeint(eq1, y0_1, x_vals)

# Plot
plt.figure()
plt.plot(x_vals, y_vals_1, label="Equation 1: dy/dx = e^(-y) sec^2(x)")
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
```

```python
# 2. Equation 2: dy/dx = ln(x) / (x * y^2), with the initial condition y(e) = 1
def eq2(y, x):
    return (np.log(x) / (x * y**2))

y0_2 = 1

# Values of x (starting from e and moving to the right)
x_vals_2 = np.linspace(np.e, 10, 100)
y_vals_2 = odeint(eq2, y0_2, x_vals_2)

# Plot 
plt.figure()
plt.plot(x_vals_2, y_vals_2, label="Equation 2: dy/dx = ln(x) / (x * y^2)")
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
```

```python
# 3. Equation 3: (sqrt(x) y') / (2 + y) = 1, rewritten as y' = sqrt(x) * (2 + y)
def eq3(y, x):
    return np.sqrt(x) * (2 + y)

y0_3 = 0
x_vals_3 = np.linspace(1, 10, 100)
y_vals_3 = odeint(eq3, y0_3, x_vals_3)

# Plot
plt.figure()
plt.plot(x_vals_3, y_vals_3, label="Equation 3: (sqrt(x) y') / (2 + y) = 1")
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
```
