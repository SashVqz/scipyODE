```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def system(vars, t):
    x, y, z = vars
    dxdt = y + z
    dydt = x + z
    dzdt = x + y
    return [dxdt, dydt, dzdt]

# Time points where we want to evaluate the solution
t = np.linspace(0, 10, 400)

# Initial conditions for x(0), y(0), z(0)
initial_conditions = [1, 0, 0]

# Solve the system of ODEs
solution = odeint(system, initial_conditions, t)

# Extract the individual solutions for x(t), y(t), z(t)
x = solution[:, 0]
y = solution[:, 1]
z = solution[:, 2]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='x(t)')
plt.plot(t, y, label='y(t)')
plt.plot(t, z, label='z(t)')
plt.title('Solution of the system of ODEs')
plt.xlabel('Time t')
plt.ylabel('Solutions x(t), y(t), z(t)')
plt.legend()
plt.grid(True)
plt.show()
```
